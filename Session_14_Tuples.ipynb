#operations of Tuples

```python
tuple1 = (1,2,3,4) #Length operation
print(len(tuple1))
```

```python
 #Delete operation
tuple1 = (1,2,3,4)
del tuple1
print(tuple1)
```

```python
# Concatenation
tuple1 = (1,2,3,4)
tuple2 = ("python","java","html")
print(tuple1 + tuple2)
```

```python
# Slicing 
tuple1 = (1,2,3,4)
print(tuple1[0:3])
```

```python
tuple1 = (1,2,3,4)
print(tuple1[::-1])
```

```python
#Looping 
tuple1 = (1,2,3,4)
for x in tuple1:
  print(x)
```

```python
tuple1 = (1,2,3,4)
i = 0
while i<len(tuple1): #0<4 #0<4 #1<4 #2<4 #3<4 #4<4 X
  print(tuple1[i]) #tuple1[0] [1] [2]  [3]
  # i = i+1 #0+1 = 1 #2 #3 #4
```

```python
tuple1 = (1,2,4,4,5) #built in function # it counts the element
print(tuple1.count(4))
```

```python
print(sum(tuple1)) #1+2+4+4+5
```
